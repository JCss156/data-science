```python

#score = [0] * len(moviesId)
#for i in range(len(moviesId)):
#    score[i] = avg_rating[i] * log(raing_count[i])
dfc['date'] = pd.to_datetime(dfc['date'])

dfus = dfc[dfc['location']=='United States']
dfaf = dfc[dfc['location']=='South Africa']

dfus = dfus.set_index('date')
dfaf = dfaf.set_index('date')

dfus = dfus.sort_index()
dfaf = dfaf.sort_index()

dfus['daily'] = dfus['total_deaths_per_million'].diff()
dfaf['daily'] = dfaf['total_deaths_per_million'].diff()

dfus['daily'] = dfus['daily'].clip(lower=0)
dfaf['daily'] = dfaf['daily'].clip(lower=0)

dfus['t'] = range(len(dfus))
dfaf['t'] = range(len(dfaf))

dfus_model = dfus[['t','daily']].dropna()
dfaf_model = dfaf[['t','daily']].dropna()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np



dfus_model = dfus[['t','daily']].dropna()
dfaf_model = dfaf[['t','daily']].dropna()

X_us = dfus_model[['t']]
y_us = dfus_model['daily']

X_af = dfaf_model[['t']]
y_af = dfaf_model['daily']

model_us = LinearRegression().fit(X_us, y_us)
model_af = LinearRegression().fit(X_af, y_af)
print(dfus_model.shape)
print(dfaf_model.shape)
```

```python
future_steps = 200

t_future_us = np.arange(len(dfus), len(dfus)+future_steps).reshape(-1,1)
pred_us = model_us.predict(t_future_us)

t_future_af = np.arange(len(dfaf), len(dfaf)+future_steps).reshape(-1,1)
pred_af = model_af.predict(t_future_af)


future_dates_us = pd.date_range(
    start=dfus.index[-1],
    periods=future_steps+1,
    freq='D'
)[1:]

future_dates_af = pd.date_range(
    start=dfaf.index[-1],
    periods=future_steps+1,
    freq='D'
)[1:]
```

```python
plt.figure(figsize=(12,6))

# real
plt.plot(dfus.index, dfus['daily'], label='US real')
plt.plot(dfaf.index, dfaf['daily'], label='AF real')

# previsão
plt.plot(future_dates_us, pred_us, '--', label='US previsão')
plt.plot(future_dates_af, pred_af, '--', label='AF previsão')

plt.legend()
plt.grid()
```
